```python
import platform
import numpy as np
import pandas as pd
import sklearn as sk
import altair as alt
import os
```

```python
teams = pd.read_csv('./datasets/2018-19 Team Stats.csv')
```

```python
teams.head()
```

```python
team1 = teams.iloc[:1]
brush = alt.selection_interval()

alt.Chart(teams).mark_point().encode(
    x='W:Q',
    y='3P:Q',
    color=alt.condition(brush, 'Team', alt.value('lightgray'))
).add_selection(
    brush
)
```

```python
team1 = teams.iloc[:1]

alt.Chart(teams).mark_bar().encode(
    x='Team',
    y='3PA',
    color=('Team')
)
```

```python
team1 = teams.iloc[:1]

points = alt.Chart(teams).mark_point().encode(
    alt.X('W', scale=alt.Scale(zero=False)),
    alt.Y('3P', scale=alt.Scale(zero=False)),
    color=alt.condition(brush, 'Team', alt.value('lightgray')),
    tooltip = [alt.Tooltip('Team'),
               alt.Tooltip('W'),
               alt.Tooltip('3P')
              ]
).add_selection(
    brush
)


bar = alt.Chart(teams).mark_bar().encode(
    y='Team',
    color='Team',
    x='3PA',
).transform_filter(
    brush
)
points & bar
```

```python
source = pd.read_csv('./datasets/2014-15 Team Stats.csv')

bars = alt.Chart(source).mark_bar().encode(
    x=alt.X('3PA', stack='zero'),
    y=alt.Y('Team'),
    color=alt.Color('W')
)

text = alt.Chart(source).mark_text(dx=-15, dy=3, color='white').encode(
    x=alt.X('3PA', stack='zero'),
    y=alt.Y('Team'),
    detail='W',
    text=alt.Text('3PA', format='.1f')
)

bars + text
```
