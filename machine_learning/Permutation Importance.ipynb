```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_csv('./datasets/2014-15 Perm.csv')
data.head()
```

```python
data = data.drop(['Team','G','MP'],axis=1)
data.head()
```

```python
data.dtypes
```

```python
data = data.drop(['3P%','2P%','FT%','FG%','FTA','3PA','2PA','FGA'],axis=1)
```

```python
data['WP'].round(decimals=0)
data['WP'] = data['WP'].astype(int)
```

```python
data.head()
```

```python
y = (data['W'])
feature_names = [i for i in data.columns if data[i].dtype in [np.float64]]
X = data[feature_names]
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
my_model = RandomForestClassifier(n_estimators=100,
                                  random_state=0).fit(train_X, train_y)
```

```python
import eli5
from eli5.sklearn import PermutationImportance
from sklearn.svm import SVC
```

```python
svc = SVC().fit(val_X, val_y)
perm = PermutationImportance(svc).fit(val_X, val_y)
eli5.show_weights(perm, feature_names=list(X.columns))
```
